####**Bar Plot Of Total Attacks Over The Time**

````python
# Chart - 1 visualization code
# Bar - Plot for the attacks per year
# Set The Plot Style
# Set The Plot Style
sns.set_style("darkgrid")

# Prepare The Data
x_year = terror['year'].unique()
y_year = terror['year'].value_counts().sort_index()

# Prepare The Figure
plt.figure(figsize=(18, 8))

# Select A Colormap
colormap = plt.cm.Reds
normalize = plt.Normalize(min(y_year), max(y_year))

# Plot The Chart
sns.barplot(x=x_year, y=y_year, palette=colormap(normalize(y_year)))

# Customize The Plot
plt.xticks(rotation=45)
plt.xlabel("Year Of Attack")
plt.ylabel("Number Of Attack")
plt.title("Bar Plot For Number of Terrorist Activities Per Year")

# Show The Plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Bar charts are effective for comparing the magnitude of different categories or groups. In the context of a terrorism dataset, each bar represents a specific year, and the height of the bar corresponds to the total number of attacks in that year. This makes it easy to compare the number of attacks across different years at a glance.

Analyzing a terrorism dataset can serve several business objectives, depending on the goals and context of the analysis. Here are some common business objectives for conducting an analysis on a terrorism dataset:

1. **Risk Assessment and Mitigation:**
   - **Objective:** Identify regions or areas with a higher risk of terrorist activities and develop strategies to mitigate these risks.
   - **Business Impact:** Enhanced security measures, improved risk management, and proactive planning to safeguard assets, personnel, and operations.

2. **Resource Allocation Optimization:**
   - **Objective:** Optimize the allocation of resources, including budget, personnel, and technology, to areas or periods with higher security risks.
   - **Business Impact:** Cost savings, more effective use of resources, and improved overall security preparedness.

3. **Strategic Decision-Making:**
   - **Objective:** Inform strategic decisions regarding market entry, expansion, or withdrawal based on the security landscape.
   - **Business Impact:** Improved strategic planning, reduced exposure to security risks, and better alignment with long-term business goals.

4. **Supply Chain Resilience:**
   - **Objective:** Enhance the resilience of the supply chain by identifying and mitigating security risks that may impact the flow of goods and services.
   - **Business Impact:** Reduced supply chain disruptions, improved operational continuity, and increased customer satisfaction.

5. **Market Presence and Reputation Management:**
   - **Objective:** Manage and safeguard the business's market presence and reputation in regions affected by terrorism.
   - **Business Impact:** Minimized reputational damage, increased customer trust, and the ability to maintain a positive market image.

6. **Humanitarian Aid Planning:**
   - **Objective:** Plan and allocate resources for humanitarian aid in regions affected by terrorism.
   - **Business Impact:** Positive contribution to affected communities, strengthened corporate social responsibility, and improved community relations.

7. **Investor Confidence and Stakeholder Communication:**
   - **Objective:** Maintain or enhance investor confidence by effectively communicating security measures and risk management strategies.
   - **Business Impact:** Positive relationships with stakeholders, sustained or improved stock prices, and continued investment support.

8. **Regulatory Compliance:**
   - **Objective:** Ensure compliance with regional and international security regulations and requirements.
   - **Business Impact:** Avoidance of legal and regulatory issues, reduced risk of fines or sanctions, and smoother business operations.

9. **Employee Safety and Well-being:**
   - **Objective:** Ensure the safety and well-being of employees by implementing measures to protect them in regions with security risks.
   - **Business Impact:** Improved employee morale, increased productivity, and reduced turnover.

10. **Community Engagement:**
    - **Objective:** Engage with local communities in regions affected by terrorism to build positive relationships and contribute to security initiatives.
    - **Business Impact:** Positive community relations, enhanced corporate social responsibility, and potential business opportunities.

These objectives highlight the importance of using terrorism data analysis to inform decision-making, enhance security measures, and contribute to the overall resilience and success of the business in regions with security challenges.
